Falling ball. Imagine that a 50 cm diameter solid aliminum (density = 2700 kg/m3) ball is dropped from rest at 10,000 m. Consider the force of gravity (F = âˆ’mg $\hat{j}$) and numerically integrate particle position and velocity versus time. For simplicity, assume constant gravity (a = 9.81 m/s2) Terminate the simulation when the ball hits the ground (y = 0). Plot position and velocity versus time with both gravity and drag included. Derive an analytical expression for the velocity as a function of height and compare to your numerical predictions.


Analytical solution: v(y) = sqrt(2g(y0 - y))

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def XtoL(x, dx, x0=0):
    return (x-x0)/dx

def gather(li, field):
    i = int(li)
    di = li - i 
    return field[i]*(1-di) + field[i+1]*di

# Set up forces
g = 9.81
rho = 2700
r = 0.5/2
m = 4/3 * rho * np.pi * r**3
F = -m*g

# Set up time steps and initial conditions
nt = int(450) # num of timesteps
dt = 1e-1 # timestep
t = np.linspace(0, nt*dt, nt)
y = 10e3
v = 0 - 0.5 * (F/m) * dt # rewind velocity
y_particle, v_particle = np.zeros(nt), np.zeros(nt)

# Solve
for k in range(nt):
    # integrate position and velocity
    v = v + F/m *dt
    y = y + v * dt
    y_particle[k] = y
    v_particle[k] = v
    
```

```python
# Plotting
y_pred = np.linspace(0, 10e3, 500)
y0 = 10e3
plt.plot(y_particle, v_particle)
plt.plot(y_pred, -np.sqrt(2*g*(y0-y_pred)), linestyle='--', linewidth = 2)
plt.ylabel('dy/dt [m/s]')
plt.xlabel('y / m')
```

Next, add the action of aerodynamic drag, $Fd = (1/2)\rho_g  C_d Av^2$. Use $C_d$ = 0.47 and assume constant air density, $\rho_g$ = 1.225 kg/m3. Plot the height and velocity versus time on the same graph as above. You should observe that the ball reaches a certain terminal velocity. Compare the simulation value to theory.

```python
# Set up forces
g = 9.81
rho = 2700
rho_g = 1.225
Cd = 0.47
r = 0.5/2
A = np.pi*r*r
m = 4/3 * rho * np.pi * r**3

Fg = -m*g

# Set up time steps and initial conditions
nt = int(450) # num of timesteps
dt = 1e-1 # timestep
t = np.linspace(0, nt*dt, nt)
y = 10e3
v = 0 - 0.5 * (Fg/m) * dt # rewind velocity
Fd = +0.5 * rho_g * Cd * A * v * v
v_terminal = np.sqrt(-Fg*2/(rho_g*Cd*A))

y_particle, v_particle = np.zeros(nt), np.zeros(nt)

# Solve
for k in range(nt):
    # integrate position and velocity
    F = Fg + Fd
    v = v + F/m *dt
    y = y + v * dt
    Fd = 0.5 * rho_g * Cd * A * v * v
    y_particle[k] = y
    v_particle[k] = v

plt.plot(t, v_particle)
plt.xlabel('time / s')
plt.ylabel('dy/dt [m/s]')
plt.axhline(-v_terminal,linestyle='--', color='red', label = 'Predicted terminal velocity')
plt.legend()
```

Surface interaction. Remove drag by setting Cd to 0

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants and parameters
g = 9.81  
rho = 2700  
rho_g = 1.225  
Cd = 0.47  
r = 0.5 / 2 
A = np.pi * r**2  
m = 4 / 3 * rho * np.pi * r**3 

Fg = -m * g  #

# Time parameters
nt = int(1e4)  # number of timesteps
dt = 0.4e-1  # timestep [s]
t = np.linspace(0, nt * dt, nt)

# Initial conditions
y = 10e3  # initial height [m]
v = 0 - 0.5 * (Fg / m) * dt  # rewind velocity
Fd = -np.sign(v) * 0.5 * rho_g * Cd * A * v**2

# Particle motion arrays
y_particle, v_particle = np.zeros(nt), np.zeros(nt)

# Simulation loop
for k in range(nt):
    # Rebound condition
    if y <= 0:
        v = -v
        Fd = -np.sign(v) * 0.5 * rho_g * Cd * A * v**2

    # Integrate position and velocity
    F = Fg + Fd
    v = v + F / m * dt
    y = y + v * dt

    # Store values
    y_particle[k] = y
    v_particle[k] = v

    # Update drag force
    Fd = -np.sign(v) * 0.5 * rho_g * Cd * A * v**2

# Compute x position (constant horizontal velocity)
ux = 10 
x_particle = ux * t

# Plot results
plt.plot(x_particle, y_particle)
plt.ylabel('y [m]')
plt.xlabel('x [m]')
plt.show()
```
